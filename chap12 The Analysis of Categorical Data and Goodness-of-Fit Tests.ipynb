# 12.1 Chi-Square Tests for Univariate Data

ùëò = number of categories of a categorical variable<br>
$ùëù_1$ = population proportion for Category 1<br>
$ùëù_2$ = population proportion for Category 2<br>
‚Ä¶‚Ä¶<br>
$ùëù_ùëò$ = population proportion for Category ùëò

- Null hypothesis $H_0$:
    - $ùëù_1$ = hypothesized proportion for Category 1<br>
    - $ùëù_2$ = hypothesized proportion for Category 2<br>
    - ‚Ä¶‚Ä¶<br>
    - $ùëù_ùëò$ = hypothesized proportion for Category ùëò
- Alternative hypothesis $H_a$: $H_0$ is not true

## Goodness-of-fit statistic
The goodness-of-fit statistic can be calculated by
$$
X^2 = \sum_{\text{all cells}} \frac{\text{(observed cell count‚àíexpected cell count)}^2}{\text{expected cell count}}
$$
where expected cell count = sample size √ó corresponding hypothesized population proportion according to the null hypothesis $H_0.$
- The value of the $ùëã^2$  statistic reflects the magnitude of the differences between observed and expected cell counts, and it can never be negative.
- When the differences are large, the value of ùëã^2  tends to be large.  This means that large values of $ùëã^2$  suggest rejection of $ùêª_0$.
- A small value of $ùëã^2$ occurs when the observed cell counts are quite similar to those expected when $ùêª_0$ is true, and would be consistent with $ùêª_0$. 

When the null hypothesis is true and the sample size is large, the behavior of $ùëã^2$ is described approximately by a chi-square distribution (https://en.wikipedia.org/wiki/Chi-squared_distribution) with $df = k-1$.

## chi-square distribution

```python
from scipy.stats import chi2
from matplotlib import pyplot as plt
import numpy as np

# set up the degrees of freedom and x-axis range in the plot

#x-axis ranges from 0 to 20 with .001 steps
x = np.arange(0, 50, 0.001)

dfs = [2,8,10,20]
for df in dfs:
    plt.plot(x,chi2.pdf(x, df=df),label= f'df={df}')
    plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper right')
```

# 10.2 Tests for Homogeneity and Independence in a Two-way Table
**Tests for homogeneity**: whether the category proportions are the same for all the populations or treatments. 
- Hypothesis:
    - $ùêª_0$: The category proportions are the same for all the populations or treatments (homogeneity of populations or treatments).
    - $ùêª_ùëé$: The category proportions are the not all same for all the populations or treatments.
- Test statistic: 
$$
X^2 = \sum_{\text{all cells}} \frac{\text{(observed cell count‚àíexpected cell count)}^2}{\text{expected cell count}}
$$
where (expected cell count)=(row marginal total)(column marginal total)/(grand total).
- Assumptions:
    - The observed counts are from a random sample.
    - The sample size is large: All expected counts are at least 5. If some expected counts are less than 5, rows or columns of the table can be combined to achieve a table with satisfactory expected counts. 
- $P$-values: When $ùêª_0$ is true and the assumptions of the $ùëã^2$ test are satisfied, $ùëã^2$ has approximately a chi-square distribution with df = (number of rows‚àí1)(number of columns‚àí1).

**Tests for Independence**: two categorical variables are independent in a population if, for each particular category of the first variable and each particular category of the second variable,
(proportion of individuals in a particular category combination)=(proportion in specified category of first variable)√ó(proportion in specified category of second variable).

- Hypothesis:
    - $ùêª_0$: The two variables are independent.
    - $ùêª_ùëé$: The two variables are not independent. 
- Test statistic: 
$$
X^2 = \sum_{\text{all cells}} \frac{\text{(observed cell count‚àíexpected cell count)}^2}{\text{expected cell count}}
$$
where the expected cell counts are estimated from the sample data (assuming that $ùêª_0$ is true) using the formula (expected cell count) =(row marginal total)(column marginal total)/(grand total).

- Assumptions:
    - The observed counts are from a random sample.
    - The sample size is large: All expected counts are at least 5. If some expected counts are less than 5, rows or columns of the table can be combined to achieve a table with satisfactory expected counts. 
- $P$-values: When $ùêª_0$ is true and the assumptions of the $ùëã^2$ test are satisfied, $ùëã^2$ has approximately a chi-square distribution with df = (number of rows‚àí1)(number of columns‚àí1).
