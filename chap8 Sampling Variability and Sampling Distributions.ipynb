# 8.1 Statistics and Sampling Variability
- **Statistic**: A number calculated from values in a sample. 
- **Sampling variability**: The observed value of a statistic depends on the particular sample selected from the population and it will vary from sample to sample. This variability is called sampling variability.
- **Sampling distribution**: The distribution that would be formed by considering the value of a sample statistic for every possible sample of a given size from a population.

# 8.2 The Sampling Distribution of a Sample Mean

## The Sampling Distribution of a Sample Mean
- $\bar{x}$: sample mean
- $n$: sample size
- $\mu$: population mean
- $\sigma$: population standard deviation
- $\mu_x$: mean of the sampling distribution of $\bar{x}$
- $\sigma_{\bar{x}}$: standard deviation of the sampling distribution of $\bar{x}$

# Properties
- $\mu_{\bar{x}} = \mu$
- $\sigma_{\bar{x}}$ = $\frac{\sigma}{\sqrt{n}}$. This formula is exact if the population is infinite and is approximately correct if the population is finite and **no more than¬†10%¬†of the population** is included in the sample.
- When the population distribution is normal, the sampling distribution of¬†$\bar{x}$ is also normal for any sample size¬†ùëõ.
- When ùëõ is sufficiently large, the sampling distribution of $\bar{x}$ is **well approximated by a normal curve**, even when the population distribution is not normal.

Reason is Central Limit Theorem, and it is safely applied when ùëõ‚â•30.

## Central Limit Theorem
When n is sufficiently large, the $\bar{x}$ distribution will be approximately normal. The standard rule of thumb is that the theorem can safely be applied when n is greater than or equal to 30.

## standardized sampling distribution
If¬†ùëõ¬†is large ($ùëõ \ge 30)$ or the population distribution is normal, the standardized variable 
$$z = \frac{\bar{x} - \mu_{\bar{x}}}{\sigma_{\bar{x}}}  = \frac{\bar{x} - \mu_{\bar{x}}}{\sigma / \sqrt{n}}$$ 
has approximately a standard normal¬†ùëß¬†distribution.


## Example

### Generate population data

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from statistics import mean,pstdev

# Population is of size 10000, randomly generated from [0,99]
np.random.seed(seed=42) # fix the random seed
population = np.random.choice(100,10000,replace=True).astype(float)
population_mean = mean(population)
# plot
plt.title("Histogram of population distribution")
plt.hist(population)
```

### Generate sampling distribution

```python
from math import sqrt

# Generate the sampling distribution of sample mean
# Generate 300 sample means.
# Each sample mean is calcuated using 50 samples sampled with replacement
sample_means = []
n_means = 300
n_samples = 30

# Generate sample means distribution
for i in range(n_means):
    # np.random.seed()
    samples = np.random.choice(population,n_samples,replace=True)
    sample_means.append(mean(samples))

# Sample mean of standard deviaiton of "the sample mean distribution"
print("Comparison of pupulation and sample means statistics:")
print("-"*20)
mean_sample_means = mean(sample_means)
print(f"Mean of the sample means distribution is: {mean_sample_means}")
print(f"Mean of population is: {population_mean}")
print("-"*20)
print(f"Population standard deviation is: {pstdev(population)}")
print(f"Population standard deviation/sqrt(n) is: {pstdev(population)/sqrt(n_samples)}")
print(f"Sample means standard deviation is: {pstdev(sample_means)}")
```

```python
# plot
plt.title("Histogram of the sampling distribution of sample mean")
sns.histplot(sample_means,kde=True)
plt.axvline(x = mean_sample_means, color = 'b', label = f'Mean of sample means: {mean_sample_means:.2f}')
plt.axvline(x = population_mean, color = 'r', label = f'Population mean: {population_mean:.2f}')
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
```

# 8.3 The Sampling Distribution of a Sample Proportion
The sample statistic for this section is the sample proportion of success: $\hat{p} = (\text{number of success in the sample})/ùëõ$ 
- $\hat{p}$: sample proportion
- ùëõ: sample size
- ùëù: population proportion
- $\mu_{\hat{p}}$: mean of the sampling distribution of $\hat{p}$
- $\sigma_{\hat{p}}$: standard deviation of the sampling distribution of $\hat{p}$

## Properties
- $\mu_{\hat{p}} =ùëù$
- $\sigma_{\hat{p}} = \sqrt{(ùëù(1‚àíùëù))/ùëõ)}$. This formula is exact if the population is infinite and is approximately correct if the population is finite and **no more than¬†10%**¬†of the population is included in the sample.
- When ùëõ is sufficiently large and ùëù is not too near 0 or 1, the sampling distribution of $\hat{p}$ is **well approximated by a normal curve**. 
    - A conservative rule of thumb is that if both $ùëõùëù \ge 10$ and $ùëõ(1‚àíùëù) \ge 10$

## Example

### Generate population data

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from statistics import mean,pstdev

# Population is of size 10000, randomly generated 
np.random.seed(seed=42) # fix the random seed
population = np.random.choice(a=[0,1],size=10000,replace=True,p=[0.3,0.7]).astype(float)
population_proportion = mean(population)
# plot
plt.title("Histogram of population distribution")
plt.hist(population)
```

### Generate sampling distribution

```python
from math import sqrt

# Generate the sampling distribution of sample mean
# Generate 300 sample means.
# Each sample mean is calcuated using 50 samples sampled with replacement
sample_proportions = []
n_means = 300
n_samples = 30

# Generate sample means distribution
for i in range(n_means):
    # np.random.seed()
    samples = np.random.choice(population,n_samples,replace=True)
    sample_proportions.append(mean(samples))

# Sample mean of standard deviaiton of "the sample mean distribution"
print("Comparison of pupulation and sample proportion statistics:")
print("-"*20)
mean_sample_proportions = mean(sample_proportions)
print(f"Mean of the sample proportions is: {mean_sample_proportions}")
print(f"Proportion of population p is: {population_proportion}")
print("-"*20)
print(f"Population standard deviation is: {pstdev(population)}")
print(f"Population standard deviation/sqrt(n) is: {pstdev(population)/sqrt(n_samples)}")
print(f"Sample standard deviation is: {pstdev(sample_proportions)}")
```

```python
# plot
plt.title("Histogram of the sampling distribution of sample proportions")
sns.histplot(sample_proportions,kde=True)
plt.axvline(x = mean_sample_proportions, color = 'b', label = f'Mean of sample proportions: {mean_sample_proportions:.2f}')
plt.axvline(x = population_proportion, color = 'r', label = f'Population mean: {population_proportion:.2f}')
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
```
