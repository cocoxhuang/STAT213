## Comparative Bar Charts (Ex3.1)
A comparative bar chart is a graph with two or more bar charts that use the same set of horizontal and vertical axes.

```python
import pandas as pd
d = {'Associate degree': [0.07, 0.06,0.016,0.25,0.46], 'Bachelors degree': [0.05,0.07,0.14,0.29,0.45]}
Education_op = pd.DataFrame(data=d, index=["strongly disagree",'disagree','neutral','agree','strongly agree'])
Education_op
# pd.read_csv(url)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = Education_op.index.values
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2,Education_op["Associate degree"], 0.4, label = "Associate degree")
plt.bar(X_axis + 0.2, Education_op["Bachelors degree"], 0.4, label = "Bachelors degree")
  
plt.xticks(X_axis, X)
plt.xlabel("Response")
plt.ylabel("Relative frequency")
plt.title("Is education worth the cost")
plt.legend()
plt.show()
```

## Segmented Bar Charts
A segmented bar chart is similar to pie chart where the slice size is proportional to the relative frequency.<br>
Instead, it uses a rectangular bar rather than a circle to represent the entire data set.

```python
df = pd.DataFrame([['A', 10, 20, 10, 26], ['B', 20, 25, 15, 21], ['C', 12, 15, 19, 6],
                   ['D', 10, 18, 11, 19]],
                  columns=['Team', 'Round 1', 'Round 2', 'Round 3', 'Round 4'])
 
# plot data in stack manner of bar type
df.plot(x='Team', kind='bar', stacked=True,
        title='Team scores')
plt.show()
```

# Displaying Small Numerical Data
- **Stem-and-Leaf Displays**:
    - Select one or more leading digits for the stem values. The trailing digits (or sometimes just the first one of the trailing digits) become the leaves. List possible stem values in a vertical column.
    - Record the leaf for every observation beside the corresponding stem value.
    - Indicate the units for stems and leaves someplace in the display.
- **comparative stem-and-leaf display**:
    - a common stem;
    - the leaves for one data set are listed to the right of the stem values and the leaves for the second data set are listed to the left

## Histrogram (Ex3.12)

```python
url = "https://github.com/cocoxhuang/STAT213/blob/main/Ex3.12.csv?raw=true"
data = pd.read_csv(url)
```

```python
plt.hist(data,rwidth=0.7)
plt.xlabel("Number of partners")
plt.ylabel("Frequency")
plt.title("Queen bee mates")
```

#### Histogram with adjusted bins
We need to group the observed data into intervals. These intervals are sometimes called class intervals. 
There are no set rules for selecting either the number of class intervals or the length of the intervals.
The quantity √(𝑛𝑢𝑚𝑏𝑒𝑟 𝑜𝑓 𝑜𝑏𝑠𝑒𝑟𝑣𝑎𝑡𝑖𝑜𝑛𝑠) is sometimes used as an estimate of an appropriate number of intervals

```python
plt.hist(data,rwidth=0.7,bins=np.arange(1,14,3))
plt.xlabel("Number of partners")
plt.ylabel("Frequency")
plt.title("Queen bee mates")
```

### Histogram Shapes
- Unimodal: A histogram is unimodal if it has a single peak.
- Bimodal: A histogram is bimodal if it has two peaks.
- Multimodal: A histogram is multimodal if it has more than two peaks.

## Time series plots
A time series plot can be constructed by thinking of the data set as a bivariate data set, where y is the variable observed and x is the time at which the observation was made. These (x, y) pairs are plotted as a scatterplot. Consecutive observations are then connected by a line segment. This helps us to see trends over time.
